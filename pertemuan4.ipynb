```python
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join(dataset, 'titanic.csv'))
df
```

```python
# Handling missing value
# find missing value
df.info()
```

```python
df.isna().sum()
```

```python
# drop missing value
data = df
data
```

```python
# how = 'all' menghapus jika semuanya itu missing value
# how = 'any' menghapus jika ada minimal satu missing value
data.dropna(axis=0, how='all', inplace=True)
data 
```

```python
# 1 special values
# 2 aggregate values (mean, mode, median etc)

data['Age'] = data['Age'].fillna(data['Age'].median())
data
```

```python
data['Fare'] = data['Fare'].fillna(data['Fare'].mode()[0])
data
```

```python
data.isna().sum()
```

```python
perc = 50
min_count =  int(((100-perc)/100)*df.shape[0] + 1)
data = data.dropna(axis=1, thresh=min_count)
data
```
