```python
import pandas as pd
import numpy as np
import os
```

```python
dataset = '/content/drive/MyDrive/Colab Notebooks/Intermediete_AI'
```

```python
df = pd.read_csv(os.path.join(dataset, 'My Uber Drives - 2016.csv'))
df
```

```python
index = ['Start Date', 'End Date', 'Category', 'Start City', 'Stop City', 'Miles', 'Purpose']
df.columns = index
df
```

```python
conditions = [
    (df['Miles'] <= 5),
    (df['Miles'] > 5) & (df['Miles'] <= 10),
    (df['Miles'] > 10)
    ]
value = ['Short Trip', 'Medium Trip', 'Long Trip']

df['Trip Category'] = np.select(conditions, value)
df
```
