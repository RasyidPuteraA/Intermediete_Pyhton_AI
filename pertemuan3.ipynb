```python
# Conditionally adding column
import numpy as np
data['Trip Category'] = np.where(data['Miles'] > 10, 'Long Trip', 'Short')
data
```

```python
data['Trip Category'].value_counts()
```

```python
data['Time Category'] = np.where(data['Start Date'] > '2016-07-01', 'New Trip', 'Old Trip')
data
```

```python
data['Time Category'].value_counts()
```

```python
# Group by Summarize
data.groupby('Start City')['Miles'].agg(['mean'])
```
