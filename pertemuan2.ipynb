```python
import numpy as np
```

```python
# append
arr = np.arange(10)
arr
```

```python
arr = np.append(arr, (10, 11, 12))
arr
```

```python
arr2 = np.array([[1,2,3],
                 [4,5,6]])
arr2
```

```python
arr2 = np.append(arr2, [[7,8,9]], axis=0)
arr2
```

```python
arr2 = np.append(arr2, [[1,1,1]], axis=0)
arr2
```

```python
arr2 = np.append(arr2, [[0], [0], [0], [0]], axis=1)
arr2
```

```python
arr2 = np.append(arr2, [[1,2,3,4]], axis=0)
arr2
```

```python
# concate
arr_a = np.array([[1,2,3],
                  [4,5,6]])
arr_b = np.array([[7,8,9],
                  [10,11,12]])

arr_c = np.concatenate((arr_a, arr_b), axis=0)
arr_c
```

```python
arr_d = np.array([[1,2,3],
                  [4,5,6]])
arr_e = np.array([[7,8,9],
                  [10,11,12]])

arr_f = np.concatenate((arr_d, arr_e), axis=1)
arr_f
```
